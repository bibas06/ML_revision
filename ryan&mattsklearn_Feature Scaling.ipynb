Feature scaling is a preprocessing step that standardizes or normalizes the numerical features in a dataset to a similar scale. Many machine learning algorithms perform better when features are on a comparable scale.

```python
#Normalisation vs Standardization
```

```python
#Normalisation=(X-MIN(X))/MAX(X)-MIN(X) Rescales features to a fixed range (e.g., 0 to 1).
#Standardization=X-MEAN(X)/STD (Transforms data to have mean = 0 and standard deviation = 1.) it is helpful when dealing with outliers.
```

```python
import pandas as pd
```

```python
df=pd.read_csv("500hits.csv",encoding='latin-1')
```

```python
df.head()
```

```python
df=df.drop(columns=['PLAYER','CS'])
```

```python
df.info
```

```python
df.describe().round(3)
```

```python
X1=df.iloc[:,0:13]
```

```python
X2=df.iloc[:,0:13]
```

```python
X1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaleStandard=StandardScaler()
```

```python
X1=scaleStandard.fit_transform(X1)
```

```python
X1=pd.DataFrame(X1,columns=['YRS','G','AB','R','H','2B','3B','HR','RBI','BB','SO','SB','BA'])
```

```python
X1
```

```python
X1.describe().round(3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaleMinMax=MinMaxScaler(feature_range=(1,2))
#feature_range=(a, b) â Transforms features to the range [a, b].
```

```python
X2=scaleMinMax.fit_transform(X2)
```

```python
X2=pd.DataFrame(X2,columns=['YRS','G','AB','R','H','2B','3B','HR','RBI','BB','SO','SB','BA'])
```

```python
X2.head()
```

```python
X2.describe().round(3)
```
