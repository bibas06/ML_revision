```python
import pandas as pd
```

```python
import numpy as np
```

```python
miles=pd.DataFrame({"farthest_run_mi":[50,62,np.nan,100,26,13,31,50]})
```

```python
miles
```

```python
miles.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean=SimpleImputer(strategy='mean').set_output(transform='pandas')
```

```python
imp_mean
```

```python
imp_mean_transform=imp_mean.fit_transform(miles)
```

```python
imp_mean_transform
```

```python
imp_mean_marked=SimpleImputer(strategy='mean',add_indicator=True).set_output(transform='pandas')
```

```python
imp_mean_marked.fit_transform(miles)
```

```python
names=pd.DataFrame({'names':["bibas","biswas","manish","snehashis","suvam",np.nan,"suresh"]})
```

```python
names
```

```python
imp_constant=SimpleImputer(strategy='constant',fill_value="anshul")
```

```python
imp_constant
```

```python
imp_constant.fit_transform(names)
```

```python
from sklearn.compose import make_column_transformer
```

```python
df=pd.DataFrame({"names":["bibas","biswas","manish","snehashis","suvam",np.nan,"suresh"],"farthest_run_mi":[50,62,np.nan,100,26,13,31]})
```

```python
df
```

```python
type(df['names'])
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct=make_column_transformer(
    (imp_constant,['names']),
    (imp_mean,['farthest_run_mi']),
    remainder='drop'
)
```

```python
ct.set_output(transform='pandas')
```

```python
df_pandas=ct.fit_transform(df)
```
