KMeans Clustering Implementation

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#from sklearn.datasets import make_blobs in order to make your own dataset
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
#In clustering algorithm ur main focus should be on X and not on y.
#centers represent numbers of centroids 
```

```python
X.shape
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
#Elbow method
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
    #wcss or SSE
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
#K-means++ techinque is used to avoid random initilization traps.Or in other words to intialize the centroids very far apart from each others so that all the data are grouped properly and classification becomes much accurate.
```

```python
y_labels=kmeans.fit_predict(X_train)
```

```python
y_labels
```

```python
y_test
```

```python
kmeans.score(X_train,y_labels)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
y_test_labels=kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test_labels)
```

```python
#knee locator
```

```python
from kneed import KneeLocator
```

```python
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
#if wcss value is decreasing use convex instead use concave in curve
```

```python
##performance metrics
#Silhouette score
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    score=silhouette_score(X_train,kmeans.labels_)
    silhouette_coefficients.append(score)
    #The silhouette score is defined only for 2 or more clusters.
    #Let's adjust the loop to start from k=2 instead of k=1.
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("silhouette coefficients")
plt.show()
```

Hierarichal Clustering Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
pca_scaled
```

```python
#Agglomerative Clustering
#To construct a dendogram
import scipy.cluster.hierarchy as sc
#plot the dendogram
plt.figure(figsize=(20,7))
plt.title("Dendograms")

#create dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))#method='ward' means eucledian distance
plt.title("Dendogram")
plt.xlabel("Sample index")
plt.ylabel("Eucledian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,linkage='ward')
    agglo.fit(X_scaled)
    score=silhouette_score(X_scaled,agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("silhouette coefficients")
plt.show()
```
