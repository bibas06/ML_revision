```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
```

```python
dict={"area":['2600','3000','3200','3600','4000'],"price":['550000','565000','610000','680000','725000']}
```

```python
df=pd.DataFrame(dict)
```

```python
df
```

```python
df.plot(x='area',y='price',kind='scatter',color='red',marker='*')
plt.xlabel('area')
plt.ylabel('price(US$)')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)#here area is feature matrix(2-D) and price is target vector
```

```python
df['price'].squeeze()
```

```python
reg.predict([[3300]])#use reshape(1,-1) to reshape the array from 1-D to 2-D
#for single prediction with multiple features X should be 2D(1,n_features)
#for multiple prediction with multiple features X should be2D(n_samples,n_features) 
```
