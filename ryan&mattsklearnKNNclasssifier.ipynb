```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("500hits.csv",encoding="latin-1")
```

```python
df.head(10)
```

```python
df=df.drop(columns=["PLAYER","CS"])
```

```python
df
```

```python
X=df.iloc[:,0:13]
```

```python
y=df.iloc[:,13]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
scaler=MinMaxScaler(feature_range=(0,1)).set_output(transform="pandas")
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
knn=KNeighborsClassifier(n_neighbors=8)
#The n_neighbors parameter in a k-Nearest Neighbors (KNN) classifier determines the number of nearest data points (neighbors) the algorithm considers when making a prediction for a new sample.
#Small n_neighbors (e.g., 1):
#High model complexity, fits training data tightly.
#Large n_neighbors (e.g., 50):
#Low model complexity, smoother decision boundaries.
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
print(y_pred)
```

```python
y_test
```

```python
knn.score(X_test,y_test)#it will check how accurate is ur test data is.
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
##[[true positive,false positive]
##[false negative,true negative]]
```

```python
cr=classification_report(y_test,y_pred)
```

```python
print(cr)
```
